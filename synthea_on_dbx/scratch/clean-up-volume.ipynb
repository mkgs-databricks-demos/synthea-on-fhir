```python
volume_path = "/Volumes/mkgs_dev/dev_matthew_giglia_synthea_data_gen/landing"
```

```python
files = dbutils.fs.ls(volume_path)
file_names = [file.name for file in files]
```

```python
file_names
```

```python
import re

# Extract unique categories from file names
categories = set()
for file_name in file_names:
	if file_name.endswith('.csv'):
		match = re.search(r'Z_(.+)\.csv$', file_name)
		if match:
			categories.add(match.group(1))

print(f"Found {len(categories)} unique categories: {sorted(categories)}")

# Create directories for each category
for category in categories:
	dir_path = f"{volume_path}/{category}"
	try:
		dbutils.fs.mkdirs(dir_path)
		print(f"Created/verified directory: {category}/")
	except Exception as e:
		print(f"Error creating directory {category}/: {e}")

# Move files to their respective directories
moved_count = 0
for file_name in file_names:
	if file_name.endswith('.csv'):
		match = re.search(r'Z_(.+)\.csv$', file_name)
		if match:
			category = match.group(1)
			source_path = f"{volume_path}/{file_name}"
			dest_path = f"{volume_path}/{category}/{file_name}"
			try:
				dbutils.fs.mv(source_path, dest_path)
				moved_count += 1
			except Exception as e:
				print(f"Error moving {file_name}: {e}")

print(f"\nSuccessfully moved {moved_count} files into their respective directories")
```
